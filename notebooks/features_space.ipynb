```python
%matplotlib inline

# scipy version 0.18.1

from __future__ import print_function
print(__doc__)

import time
import numpy as np
import pickle
from scipy.io import loadmat
from matplotlib import pyplot as plt
```

```python
f = open('features_ordered.p', 'rb')
features = pickle.load(f)
f.close()
```

```python
C = np.corrcoef(features) #- np.eye(C.shape[0])
```

```python
fig = plt.figure(figsize=(10,10))
plt.imshow(C,cmap='bone',interpolation=None)
plt.show()
fig.savefig('features_space.png', dpi=fig.dpi)
```
